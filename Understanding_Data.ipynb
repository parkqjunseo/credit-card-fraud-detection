# Credit Card Fraud Detection(신용카드 사기 탐지)

creditcard.csv 데이터가 어떻게 구성되어있는지 확인하는 코드입니다.

```python
import os, pandas as pd, numpy as np, matplotlib.pyplot as plt
from pathlib import Path

csv_path = Path('./creditcard.csv')

if not csv_path.exists():
    raise FileNotFoundError(
        "creditcard.csv가 디렉토리에 없습니다."
    )

df = pd.read_csv(csv_path)
print(df.shape, df.columns.tolist()[:10], '...')
df.head()
```

## 1. 기본 정보

```python

df.info()
```

## 2. Label 분포

```python
class_counts = df['Class'].value_counts().sort_index()
print(class_counts)
# Bar plot (matplotlib only, single figure)
import matplotlib.pyplot as plt
fig = plt.figure()
plt.bar(class_counts.index.astype(str), class_counts.values)
plt.title('Class distribution (0=non-fraud, 1=Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

## 3. Amount & Time 분포

```python

# Amount(유로 금액) 분포 표
fig = plt.figure()
plt.hist(df['Amount'].values, bins=50)
plt.title('Amount distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()

# Time(거래 시간) 분포 표
fig = plt.figure()
plt.hist(df['Time'].values, bins=50)
plt.title('Time distribution (seconds since first txn)')
plt.xlabel('Time (s)')
plt.ylabel('Frequency')
plt.show()
```

## 4. 상관행렬 (V1~V28, Amount, Class)

```python

cols = [c for c in df.columns if c.startswith('V')] + ['Amount', 'Class']
corr = df[cols].corr()
fig = plt.figure(figsize=(6,5))
plt.imshow(corr, interpolation='nearest')
plt.title('Correlation matrix')
plt.colorbar()
plt.xticks(range(len(cols)), cols, rotation=90)
plt.yticks(range(len(cols)), cols)
plt.tight_layout()
plt.show()
```

## 5. 데이터가 정상적인지 체크(Sanity Check)

```python

print('Nulls per column:\n', df.isnull().sum().sort_values(ascending=False).head())
print('Descriptive stats:\n', df.describe().T.head(10))
```
